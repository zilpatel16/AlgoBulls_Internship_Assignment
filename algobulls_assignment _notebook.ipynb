## 1. Defining a Class 'ScriptData', required methods namely 'fetch_intraday_data', 'convert_intraday_data' and additional methods for overloading


```python

import requests
from datetime import datetime

class ScriptData:
    def __init__(self, api_key='RU87PW8XFVI2E05Q'):
        self.api_key = api_key
        self.data = {}
    
    #function to fetch data through AlphaVantage API
    def fetch_intraday_data(self, script):
        url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={script}&interval=60min&apikey={self.api_key}'
        response = requests.get(url)
        if response.status_code == 200:
            self.data[script] = response.json()['Time Series (60min)']
    
    #function to convert fetched json data into pandas dataframe
    def convert_intraday_data(self, script):
        if script not in self.data:
            print(f"No intraday data found for {script}")
            return None
        df = pd.DataFrame.from_dict(self.data[script], orient='index')
        df.reset_index(inplace=True)
        df.columns = ['timestamp', 'open', 'high', 'low', 'close', 'volume']
        
        #stripping time stamp that was in string format and converting integer type to float
        
        df['timestamp'] = df['timestamp'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')) #rev
        df[['open', 'high', 'low', 'close', 'volume']] = df[['open', 'high', 'low', 'close', 'volume']].astype(float)
        return df

    def __getitem__(self, script):
        if script not in self.data:
            print(f"No data found for {script}")
            return None
        return self.data[script]

    def __setitem__(self, script, data):
        self.data[script] = data

    def __contains__(self, script):
        return script in self.data

    
```

```python
script_data = ScriptData() #creating an instance
script_data.fetch_intraday_data('MSFT') #fetching the data
script_data.convert_intraday_data('MSFT') #converting the data into a dataframe
```

## 2. Creating indicator1 function

```python
#creating indicator1 function as instruted
def indicator1(df, timeperiod):
    ma = df['close'].rolling(timeperiod).mean() #using rolling function to calculate moving avergae
    output_df = pd.DataFrame({'timestamp': df['timestamp'], 'indicator': ma})
    return output_df
```

## Defining the Strategy class and get_signals method

```python
class Strategy:
    def __init__(self, api_key):
        self.api_key = api_key
        self.sd = ScriptData(api_key)
        

    def get_signals(self, script):
        # Fetch intraday historical data
        dfi = self.sd.fetch_intraday_data(script)
        df = self.sd.convert_intraday_data(script)
        close_data = df['close']

        # Compute indicator data using indicator1 function
        indicator_data = indicator1(df, timeperiod=5)['indicator']

        # Generate signals
        signals = pd.DataFrame({'timestamp': df.timestamp})
        signals['signal'] = 'NO_SIGNAL'
        signals.loc[indicator_data > close_data, 'signal'] = 'BUY'
        signals.loc[indicator_data < close_data, 'signal'] = 'SELL'
        
        #excluding signals column where there are no signals 
        
        signals_final=signals[signals['signal']!= 'NO_SIGNAL']
        return signals_final
```
